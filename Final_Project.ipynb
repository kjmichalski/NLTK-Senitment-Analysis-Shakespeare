```python
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('english'))
```

```python
f = open("shakespeare_sonnets.txt", "r", encoding = 'utf-8')
```

```python
contents = f.read()
```

```python
contents
```

```python
listSonnets = contents.split('#')
```

```python
listSonnets
```

```python
sentences = nltk.sent_tokenize(contents)
```

```python
print(len(sentences))
```

```python
lengthSonnets = [len(sonnet) for sonnet in listSonnets]
```

```python
print(lengthSonnets)
```

```python
sonnetIndex = range(1,len(lengthSonnets)+1)
```

```python
dictionary = zip(sonnetIndex, lengthSonnets)
```

```python
for (x,y) in dictionary:
    print(x,y)
```

```python
lengthSonnetsWords = [len(sonnet.split()) for sonnet in listSonnets]
```

```python
lengthSonnetsWords
```

```python
tokens = word_tokenize(contents)
```

```python
tokens
```

```python
for token in tokens:
    if token in stop_words:
        tokens.remove(token)
```

```python
freq = nltk.FreqDist(tokens)
```

```python
freq
```

```python
for key,val in freq.items():
    print(val,key)
```

```python
def sonnetID(item):
    return ("Sonnet Number " + str(item) + " has " + str(lengthSonnets[item-1]) + " characters "
           + "and " + str(lengthSonnetsWords[item-1])+ " words")
```
